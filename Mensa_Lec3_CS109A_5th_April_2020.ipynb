```python
# I create a dataset as list, a list of column name (header), and insert ID header to beginning

dataset = []

f = open("data/top50.csv", encoding = 'utf8')
column_names = f.readline().strip().split(",")[1:]
column_names.insert(0,"ID")

# Process every line in f
# split them to a list of attributes
# use zip function to match them by position a[1] - b[1], a[2] - b[2],...
# then I create dictionary, whose keys is column name, value is attribute
# finally, I  append them to my list dataset
# ?: why line begins from second row in my file ?

for line in f:
    print(line)
    attributes = line.strip().split(",")
    a = dict(zip(column_names, attributes))
    dataset.append(a)
    #a = dict(zip(column_names, attributes))
    
```

```python
for song in dataset:
    if int(song['Length']) > 240:
        print("{0} - performed by {1}".format(song['TrackName'],song['ArtistName']))
```

```python
# I use operator module to sort my dataset
# I use itemgetter() to get 'popularity' as my key to sort
# !!! : attention to song 40th '[damn],...', its attributes were incorrectly arranged
# !!! : reason : ',' in its name
import operator
sorted(dataset,key= operator.itemgetter('Popularity'),reverse = True)
```
